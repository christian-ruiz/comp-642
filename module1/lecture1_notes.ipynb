## Machine Learning Flow

0. understand the problem

1. get the data

2. clean, prepare & manipulate the data

3. train model

4. test data

5. improve

## When to use ML

- human expertise does not exist
    - navigating mars

- humans are unable to explain their expertise
    - speech recognition

- solution changes in time
    - routing on a computer network, cybersecurity

## Relation to other fields of study

- we want to optimize a performance criterion using example data or past experience so a computer can do what we do as thinking entities
    - role of statistics: inference from a sample
    - role of computer science: efficient algorithms

## Model selection & generalization

- learning is an ill-posed problem; data is not sufficient to find a unique solution

- the need for inductive bias, assumptions about the hypothesis space

- generalization: how well a model performs on new data
    - overfitting: hypothesis space is more complex than the class or function
    - underfitting: hypothesis space is less complex than the class or function

## Categories of Learning

- supervised learning
    - input data includes label (desired solution) which is used to "train" the system

- unsupervised learning
    - input data is not labeled

- semisupervised learning
    - some input data is labeled and is not

- reinforcement learning
    - system interacts and receives rewards or penalties based on decision to determine a strategy (called a policy)

## Supervised Learning

- prediction of future cases
    - use the rule to predict the output for future inputs

- knowledge extraction
    - the rule is easy to understand

- compression
    - the rule is simpler than the data it explains

- outlier detection
    - exceptions that are not covered by the rule

## unsupervised learning

- learning 'what normally happens'

- no output

- clustering
    - grouping similar instances

- example applications
    - customer segmentation in crm
    - image compression
        - color quantization
    - Bioinformatics:
        - learning motifs

## Reinforcement learning

- learning a policy
    - a sequence of outputs

- no supervised output but delayed reward

- credit assignment problem

- game playing

- robot in a maze

- multiple agents, partial observability

## ML problem types

- classification
    - assign a category to each item (document classification)

- regression
    - predict a real value for each item (prediction of stock values, economic variables)

- ranking
    - order items according to some criterion (relevant web pages returned to a search engine)

- clustering
    - partition data into homogenous regions ( analysis of very large data sets)

- association
    - discover relationships between items in large data sets (market basket analysis)

- dimensionality reduction
    - find lower-dimensional manifold preserving some properties of the data

## clustering

- objective of clustering is to find different groups within the elements in the data

- clustering algorithms find the structure in the data so that elements of the same cluseter (or group) are more similar to each other than to those from different clusters

## dimensionality reduction

- feature selection
    - finds a subset of the original set of variables, or features, to get a smaller subset which can be used to model the problem

- feature extraction
    - this reduces the data in a high dimensional space to a lower dimension space

## Why deep learning

- other ml techniques require human analysis and decisions

- deep learning uses multiple layers to progressively extract high level features from raw input

- for example, in image processing, lower layers may identify edges, while higher layers may identify human-meaningful items such as digits/letters or faces
